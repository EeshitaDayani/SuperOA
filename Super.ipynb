```python
import pandas as pd
from io import StringIO
import csv
import numpy as np
```

```python
data = StringIO('Airline Code;DelayTimes;FlightCodes;To_From\nAir Canada (!);[21, 40];20015.0;WAterLoo_NEWYork\n<Air France> (12);[];;Montreal_TORONTO\n(Porter Airways. );[60, 22, 87];20035.0;CALgary_Ottawa\n12. Air France;[78, 66];;Ottawa_VANcouvER\n""".\\.Lufthansa.\\.""";[12, 33];20055.0;london_MONTreal\n')
df = pd.read_csv(data, sep=";")
df
```

```python
# 1
import math

ref_idx = 0

# find first non-Nan value
for i in range(len(df)):
    if math.isnan(df.at[i, 'FlightCodes']) == 0:
      ref_idx = i
      break

val = df.at[ref_idx, 'FlightCodes']

for i in range(ref_idx, len(df)):
  df.at[i, 'FlightCodes'] = val + i * 10

for i in range(0, ref_idx):
  df.at[i, 'FlightCodes'] = val - (ref_idx - i) * 10

df = df.astype({'FlightCodes':'int'})
df
```

```python
# 2a
df[['To', 'From']] = df["To_From"].apply(lambda x: pd.Series(str(x).split("_")))
df
```

```python
# 2b
df['To'] = df['To'].str.capitalize()
df['From'] = df['From'].str.capitalize()
df
```

```python
# 3
df["Airline Code"] = df['Airline Code'].str.replace('[^A-Za-z\s]+','')
df
```
